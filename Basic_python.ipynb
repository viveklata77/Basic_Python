**7. What is the difference between “==” and “is” operators in Python?**

Ans: The 'is' operator in Python is used to check if two variables point to the same object. Unlike the '==' operator, which checks if the values of two objects are equal, the 'is' operator goes one step further to ensure that they are, in fact, the exact same object.

 eg.:
  == (Equality Operator):
Compares the values of two objects.
Returns True if the objects have the same value, and False otherwise.
Example: 1 == 1 is True, 1 == 2 is False, and "hello" == "hello" is True.

  is (Identity Operator):
Compares the memory addresses of two objects.
Returns True if both variables point to the same object in memory, and False otherwise.

Example: a = 1; b = 1; a is b might be True (depending on Python's internal caching), but a = [1]; b = [1]; a is b is False because they are different lists, even if they have the same content.

**10. Write code to demonstrate type casting with list elements.**

```python
# Example list with different data types
l = [10, "20", 30.5, "40.7", True]
l1 = []
for item in l:
    try:
      l1.append(int(item))
    except ValueError:
        try:
            l1.append(float(item))
        except ValueError:
            if isinstance(item, bool):
                l1.append(int(item))
                l1.append(item)
print("Original List:", l)
print("Casted List:", l)
```

**13. Write a Python program to find the sum of all even numbers between 1 and 50.**

```python
def even_sum(n):
    sum = 0
    for i in range(0,n+1,2):
        sum = sum + i
    return sum
print(even_sum(50))
```

**15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.**

```python
fact = 1
i = 1
num = int(input("Enter a number: "))
if num < 0:
    print("Sorry, factorial does not exist for negative numbers")
else:
    while i <= num:
        fact = fact*i
        i = i + 1
    print("The factorial of", num, "is", fact)
```
